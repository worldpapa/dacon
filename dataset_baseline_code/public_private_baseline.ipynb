# DataFrame

```python
import pandas as pd
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning) # FutureWarning 제거

# csv 형식으로 된 데이터 파일을 읽어옵니다.
df = pd.read_csv('car_scrape.csv')
#submission = pd.read_csv('data/sample_submission.csv')
df.head()
```

```python
df = df.reset_index().rename({'index':'id'}, axis = 'columns')
```

```python
df.head()
```

```python
df.columns
```

```python
df = df[['id', 'title', 'odometer', 'location', 'isimported', 'engine',
       'transmission', 'fuel', 'paint', 'year','price']]
df.head()
```

```python
df = df.rename({'price':'target'}, axis = 'columns')
df.head()
```

```python
df.to_csv('data/df.csv', index=False)
```

# Train, Test

```python
from sklearn.model_selection import train_test_split

train, test = train_test_split(df, test_size=0.3, stratify = df['transmission'] )
```

```python
train.columns
```

```python
print(train.shape, test.shape)
```

```python
train = train.drop('id', axis = 1)
test = test.drop('id', axis = 1)
```

```python
train = train.reset_index().drop('index', axis = 1).reset_index().rename({'index':'id'}, axis = 'columns')
test = test.reset_index().drop('index', axis = 1).reset_index().rename({'index':'id'}, axis = 'columns')
```

```python
train
```

```python
test_total = test.copy()
test = test.drop('target', axis = 1)
```

```python
train.to_csv('data/train.csv', index=False)
test.to_csv('data/test.csv', index=False)
test_total.to_csv('data/test_total.csv', index=False)
```

# Test - Public, Private

```python
from sklearn.model_selection import train_test_split

public_answer, private_answer = train_test_split(test_total[['target']], test_size=0.5) #, stratify = test_total['target'] )
```

```python
private_answer.to_csv('data/private_answer.csv', index=False)
public_answer.to_csv('data/public_answer.csv', index=False)
```
